<h3> 3. Check whether all the zones generate the same amount of sales.

H0 -μA = μB = μC = μD(where μ = mean)

Ha - At least one difference among the mean sales.

```python
from scipy.stats import f_oneway
f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'],df['Zone - D'])
```

The F test statistic is 5.672056106843581 and the p-value is 0.0011827601694503335 then the p-value is less than .05, we reject the null hypothesis.Then we can said that there is difference  in sales of the four zones

```python
df.boxplot(column=['Zone - A','Zone - B','Zone - C','Zone - D'], figsize=(10, 6), showmeans=True)
```

The above figure shows that all the zones sales are different and also the green line shows the median and the  green small triangle shows the mean.
